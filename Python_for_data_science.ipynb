```python
mi_lista = [1,2,3,4]
```

```python
def cuadrado(n):
    return n*n
```

```python
def calcular_cuadrado(numero):
  for numeros in numero:
    resultado = cuadrado(numeros)
    print(f'El cuadrado de {numeros} es {resultado}')
```

```python
calcular_cuadrado(mi_lista)
```

```python
### LIMPIEZA DE DATOS

data = {"Id_Producto":[1001,1002,1003,1003]
        ,"Cantidad_Vendida":[30, None,25,25]
        ,"Precio":[20.5,15.0,None,22.5]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_eliminados = df.dropna()
df_eliminados
```

```python
valores_nuevos = {"Cantidad_Vendida":25,
                  "Precio":df["Precio"].mean()}
df_rellenados = df.fillna(valores_nuevos)
df_rellenados
```

```python
df_rellenados["Cantidad_Vendida"] = df_rellenados["Cantidad_Vendida"].astype(int)
df_rellenados
```

```python
df_unicos = df_rellenados.drop_duplicates(subset='Id_Producto')
df_unicos
```
